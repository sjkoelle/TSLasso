Our data lies in a grid on a cylindrical surface via the equations
\begin{eqnarray*}
z_1 &=& x_1\cos(x_2) \\\\
z_2 &=& x_1\sin(x_2) \\\\
z_3 &=& x_2
\end{eqnarray*}
which we then randomly rotate in 50 dimensions. Note that we can find the tangent space analytically by SVD.

```python
#nx = 2
#nz = 50
nz = 50
zs = np.zeros((n, nz))
dz = np.zeros((n,nz, nx))
for i in range(n):
    pos = torch.tensor(xs[i], requires_grad=True)
    z_temp = pos[0]*torch.cos(pos[1])
    zs[i,0] = z_temp
    z_temp.backward()
    dz[i,0,:] = pos.grad.detach().numpy()
    pos.grad.data.zero_()
    pos = torch.tensor(xs[i], requires_grad=True)
    z_temp = pos[0]*torch.sin(pos[1])
    zs[i,1] = z_temp
    z_temp.backward()
    dz[i,1,:] = pos.grad.detach().numpy()
    pos.grad.data.zero_()
    pos = torch.tensor(xs[i], requires_grad=True)
    z_temp = pos[1]
    zs[i,2] = z_temp
    z_temp.backward()
    dz[i,2,:] = pos.grad.detach().numpy()
    pos.grad.data.zero_()
```

which we then randomly rotate in 50 dimensions.  This defines a 2 dimensional manifold.
Our dictionary is
\begin{eqnarray*}
\{x_1, x_2, x_1^2, x_2^2, sin(x_1), cos(x_1), sin(x_2), cos(x_2), x_1 x_2\}
\end{eqnarray*}
as well as the tangent coordinates to the manifold itself.

Consider a single flasso regression problem
\begin{eqnarray*}
y_i \sim x_i \beta
\end{eqnarray*}
and the tangent problem
\begin{eqnarray*}
I \sim x_i^* \beta^*
\end{eqnarray*}
where $x_i^*$ are the partial derivatives w.r.t. some manifold defined as $z_j (x)$. We can't simply invert the partial derivates unless the $z_j$ are expressable in terms of single $x^k$.  

The gradients $\nabla_x z \in \mathbb R^{K \times J}$ define coordinates for tangent spaces in $\mathbb R^K$ and $\mathbb R^J$ via the SVD $\nabla_x z = U \Sigma V^T$.  We call the dimension of these tangent spaces $d$. We can then define the pseudoinverse $dz_*^{-1} = U \Sigma^{-1*} V^T$, and compute $dz_*^{-1} \nabla_x f_p = \nabla_z f_p $.

Now we can use the fact that $y_i = \nabla_x z_j$, and $\nabla_z f_p = x_i^*$, where we have subsumed the sample point specific $i$ index.

\begin{eqnarray*}
\beta = \arg \min \| \nabla_x z_j - \nabla_x f_p  \beta \|_F^2 = \arg \min \| U \Sigma V^T - \nabla_x f_p  \beta \|_F^2  \\\\
\beta^* = \arg \min \| I_d - dz_*^{-1}  \nabla_x f_p \beta \|_F^2 = \arg \min \| I_d - U \Sigma^{-1} V^T  \nabla_x f_p \beta \|_F^2 \\\\
\end{eqnarray*}
From this, we can readily see that the two methods extremely similar.  They are equivalent if $J = K$ and $dz$ is an isometry, and also seemingly in more general settings (like when $J = K$ and $dz$ is full rank).

For example,
\begin{eqnarray*}
z_1 &=& x_1^2 \\\\
z_2 &=& sin(x_2)
\end{eqnarray*}
With flasso, we would recover partial derivatives $2x_1$, $\cos(x_2)$. With tangent space lasso, we have $\frac{\partial x_1}{\partial z_1} = \frac{1}{2x_1}, \frac{\partial x_2}{\partial z_2} = \frac{1}{\cos(x_2)}$. Thus, we recover the same partial derivatives as in the case of flasso.
