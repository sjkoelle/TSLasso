```python
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD
cov = np.identity(20)
cov[0,1] = .9
cov[1,0] = .9
cov[19,19] = 5
test_data = np.random.multivariate_normal(np.zeros(20), cov, 100)
test_data_cent = test_data - test_data.mean(axis = 0)
```

```python
# def pca(X):
#   # Data matrix X, assumes 0-centered
#   n, m = X.shape
#   # Compute covariance matrix
#   C = np.dot(X.T, X) / (n-1)
#   # Eigen decomposition
#   eigen_vals, eigen_vecs = np.linalg.eig(C)
#   # Project X onto PC space
#   X_pca = np.dot(X, eigen_vecs)
#   return(X_pca)
```

```python
# def svd(X):
#   # Data matrix X, X doesn't need to be 0-centered
#   n, m = X.shape
#   # Compute full SVD
#   U, Sigma, Vh = np.linalg.svd(X, 
#       full_matrices=False, # It's not necessary to compute the full matrix of U or V
#       compute_uv=True)
#   # Transform X with SVD components
#   X_svd = np.dot(U, np.diag(Sigma))
#   return(X_svd)
```

```python
pca = PCA(n_components=5)
pca.fit(test_data)
test_data_pca = pca.transform(test_data)
svd = TruncatedSVD(n_components=5)
svd.fit(test_data_cent)
test_data_svd = svd.transform(test_data_cent)
test_data_svd_comptrans = np.matmul(svd.components_, test_data_cent.transpose()).transpose()
test_data_pca_comptrans = np.matmul( pca.components_, test_data_cent.transpose()).transpose()
```

```python
print(np.linalg.norm(pca.components_, axis = 1))
print(np.linalg.norm(svd.components_, axis = 1))
print(pca.singular_values_)
print(svd.singular_values_)
print(test_data_pca[0] - test_data_svd[0])
print(test_data_svd[0] - test_data_svd_comptrans[0])
print(test_data_pca_comptrans[0] - test_data_pca[0])
print(pca.singular_values_**2 / 99)
print(pca.explained_variance_)
```
