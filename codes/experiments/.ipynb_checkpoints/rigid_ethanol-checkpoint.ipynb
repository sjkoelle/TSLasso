```python
def create_vector_field(loc, vec, dim=2):
    return [loc[ii] for ii in range(dim)] + [vec[ii] for ii in range(dim)]


def plot_vec_field(embedding, vectors, indices, dimension=2, color_start_ix=1,
                   namelist=None, **kwargs):
    num_vector_set = len(vectors)
    if namelist is None:
        namelist = ['vec field {}'.format(i+1) for i in range(num_vector_set)]
    for ix_ in indices:
        vec_fields = map(
            lambda x: create_vector_field(embedding[ix_], x[ix_],
                                          dimension), vectors)
        color_x = color_start_ix
        for color_x, vec_field, name in zip(range(
            color_start_ix, color_start_ix + num_vector_set),
            vec_fields, namelist):
            ax = plt.gca()
            ax.quiver(*vec_field, color=color_hex(color_x),
                      label=name if ix_ == 0 else None, **kwargs)
```
